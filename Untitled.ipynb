```python
### Code to extract  .nwo files from folders and subfolder 
import os
import shutil
import platform 

def find_files(directory_path):
    os_name = platform.system()
    for root, dirs, files in os.walk(directory_path):
        for dir in dirs:
            if os_name == "Windows":
                subfolder_path = root + dir    
            elif os_name == "Linux":
                subfolder_path = root + '/'+ dir
            for sub_root, sub_dirs, sub_files in os.walk(subfolder_path):
                for file in sub_files:
                    file_path = os.path.join(sub_root, file)
                    yield file_path

# Define the directory path
directory_path = "data\\"

def generate_custom_file_name(file_path,components):
    # Extract relevant components from the file path
    
    mod=components[-5][0:4]
    monomer_type = components[-4].split("_")[-3]  # Extract monomer type from the third last component
    method = components[-3][0:4]  # Extract method from the second last component
    molecule_name = components[-2].split("_")[-1]  # Extract molecule name from the second last component
    product_name = components[-1].split("_")[-1]  # Extract product name from the last component
    
    # Construct the custom file name
    custom_file_name = f"{mod}_{monomer_type}_{method[:3]}_{molecule_name}_{product_name.split('.')[0]}"
    return custom_file_name
destination_directory="xyz files"
destination_directory2="smiles2xyz"
# Loop through files in subfolders and find those with the specified extension
l=[]
for file_path in find_files(directory_path):
    print(file_path)
    os_name = platform.system()
    if os_name == "Windows":
        components = file_path.split("\\")
        
    elif os_name == "Linux":
        components = file_path.split("/") 
    # Check if the file type is "nwo"
    if file_path.endswith(".nwo") and len(components)>=5 and "log" not in file_path:
        # Extract the file name
        file_name = os.path.basename(file_path)
        custom_file_name = generate_custom_file_name(file_path,components)
        print(custom_file_name)
        print(file_path)
        custom_file_name = custom_file_name+ ".nwo"  # Add the file extension
        destination_path = os.path.join(destination_directory, custom_file_name)
        l.append([custom_file_name,file_path])
        shutil.copy(file_path, destination_path)
    elif file_path.endswith(".nwo") and len(components)>=5 and "log" in file_path:
        # Extract the file name
        file_name = os.path.basename(file_path)
        custom_file_name = generate_custom_file_name(file_path,components)
        print(custom_file_name)
        print(file_path)
        custom_file_name = custom_file_name+".log"+ ".nwo"  # Add the file extension
        destination_path = os.path.join(destination_directory, custom_file_name)
        l.append([custom_file_name,file_path])
        shutil.copy(file_path, destination_path)
    elif file_path.endswith(".xyz") and len(components)>=5:
        file_name = os.path.basename(file_path)
        custom_file_name = generate_custom_file_name(file_path,components)
        print(custom_file_name)
        custom_file_name = custom_file_name+ ".xyz"  # Add the file extension
        destination_path = os.path.join(destination_directory2, custom_file_name)
        l.append([custom_file_name,file_path])
        shutil.copy(file_path, destination_path)
print("done")
```

```python
import os
import numpy as np
import pandas as pd
from rdkit import Chem
import xyz2mol as x2m
from rdkit import Chem
from rdkit.Chem import AllChem

class DataExtractor:
    def __init__(self, folder_path, repo_path):
        self.folder_path = folder_path
        self.repo_path = repo_path

    def get_files_in_folder(self):
        files = os.listdir(self.folder_path)
        return files

    def read_first_line(self, file_path):
        with open(file_path, 'r') as file:
            first_line = file.readline().strip()
        return first_line

    def extract_data(self, folder_path, file_name):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r',encoding='utf-8') as file:
            lines = file.readlines()
            natom = None
            denergy = None
            nelectrons= None
            for line in lines:
                if "No. of atoms" in line:
                    natom = line.split()[-1]
                elif "No. of electrons " in line:
                    nelectrons = line.split()[-1] # Extracting 3 characters after "No of atoms"
                elif "Total DFT energy" in line:
                    denergy = line.split()[-1]  # Extracting data after "Total DFT energy"
            data = []
            with open(file_path, 'r', encoding='utf-8') as file:
                lines = file.readlines()
                start_reading = False
                for line in lines:
                    if "No.       Tag          Charge          X              Y              Z" in line:
                        start_reading = True
                        continue
                    if start_reading:
                        parts = line.split()
                        if len(parts) == 6:  # Assuming each line has exactly 6 columns
                            data.append(parts)
                        if len(data) == float(natom)+1:  # Read two lines after encountering the header
                            break
            df = pd.DataFrame(data, columns=["No.", "Tag", "Charge", "X", "Y", "Z"])
            #df.drop(0,inplace=True)

        return df,natom,nelectrons, denergy
    
    def count_number_pairs(self,string):
        stack = []  # Initialize stack
        pair_count = 0  # Initialize count for number pairs

        # Iterate through the characters in the string
        for char in string:
            if char.isdigit():
                if stack and stack[-1] == char:
                    stack.pop()  # Remove top digit from stack
                    pair_count += 1  # Increment count for number pairs
                else:
                    stack.append(char)  # Append current digit to stack

        return pair_count
    def generate_fingerprint(self,smiles_code):
        mol = Chem.MolFromSmiles(smiles_code)
        if mol is not None:
            fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, 2)
            # Convert fingerprint to numpy array
            features = np.array(fingerprint)
            return features
        else:
            return None
    def extracter_smiles(self,folder_path_smiles):
            data=[]
            for filename in os.listdir(folder_path_smiles):
                if filename.endswith(".xyz"):  # Assuming files have .txt extension
                    file_path = os.path.join(folder_path_smiles, filename)
                    print(file_path)
                    with open(file_path, 'r') as file:
                        lines = file.readlines()

                        if "SMILES" in lines[1]:

                            smiles_code = lines[1].strip().split(":")[1]
                        else:
                            smiles_code="O=C1NC2=C(N=CS2)S1"
                        digits_list = []  # Initialize list to store digits
                        pair_count=0
                    # Iterate through the characters in the string
                        for char in str(smiles_code):
                            if char.isdigit():
                                if char in digits_list:
                                    digits_list.remove(char)  # Remove the digit from the list
                                    pair_count += 1  # Increment count for number pairs
                                else:
                                    digits_list.append(char)  # Add the digit to the list

                        num_atoms = int(lines[0].strip())  # Extract number of atoms from the first line
                        data.append({"Filename": filename, "SMILES": smiles_code, "Num_Atoms": num_atoms,"nrings":pair_count})
            return data
    def process_files(self):
        data = np.empty((0, 12))
        file_names = self.get_files_in_folder()

        for file_name in file_names:
            df, natom, nelectrons, denergy = self.extract_data(self.folder_path, file_name)
            da=pd.Series(df.drop('No.',axis=1).value_counts('Tag')).reindex()
            element_order = ['C', 'H', 'S', 'N', 'O', 'Misc']
            for i, element in enumerate(element_order):
                if element is not None:
                    if element in da.index:
                        globals()[element]= da[element]
                    else:
                        globals()[element]=0
            
            #number_pairs_count = self.count_number_pairs(c)
            row = np.array([[file_name,natom, nelectrons, denergy, "smiley code", C,H,S,N,O,Misc,"numberofrings"]])
            data = np.append(data, row, axis=0)
            array = np.zeros((1, len(element_order)), dtype=int)

        return data
    def merger(self):
        folder_path_smiles = r"smiles2xyz"
        df1 = pd.DataFrame(self.extracter_smiles(folder_path_smiles))
        df2=pd.DataFrame(self.process_files())
        df2.columns=["Filename","natom", "nelectrons", "denergy", "smiley code", 'C',"H","S","N","O","Misc","numberofrings"]
        df1['Modified_File_name'] = df1['Filename'].str.split('.').str[0]
        df2['Modified_File_name'] = df2['Filename'].str.split('.').str[0]
        merged_df = pd.merge(df1, df2, on='Modified_File_name', how='inner')

        # Drop the temporary 'Modified_File_name' column
        merged_df.drop(columns=['Modified_File_name',"smiley code",'numberofrings'], inplace=True)
        merged_df.dropna(subset=['denergy'],inplace=True)
        merged_df.reset_index(inplace=True)
        merged_df['Fingerprint'] = merged_df['SMILES'].apply(self.generate_fingerprint)
        print(merged_df['Fingerprint'].iloc[0])
        num_bits = merged_df['Fingerprint'].iloc[0].shape[0]

        # Create a DataFrame to store individual features
        features_df = pd.DataFrame(merged_df['Fingerprint'].tolist(), columns=[f'Feature_{i+1}' for i in range(num_bits)])
        merged_df = pd.concat([merged_df, features_df], axis=1)
        merged_df.drop(["index","Filename_x","Filename_y"],axis=1,inplace=True)
        return merged_df
    def run(self):
        return self.merger()

folder_path = r'xyz files'
repo_path = r'git_repo'
extractor = DataExtractor(folder_path, repo_path)
df=pd.DataFrame(extractor.run())
excel_file="df_trial.xlsx"
#excel_file= input("excel_file_name")
if os.path.exists(excel_file):
    # If the file exists, delete it
    os.remove(excel_file)
df.to_excel(excel_file,index=False)
df
```

```python
# Function to plot actual vs predicted values
def plot_actual_vs_predicted(y_true, y_pred, model_name):
    plt.figure(figsize=(8, 6))
    plt.scatter(y_true, y_pred, alpha=0.5)
    plt.plot(y_true, y_true, color='red', linestyle='--')  # Straight line representing perfect predictions
    plt.title(f"{model_name} - Actual vs Predicted")
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.grid(True)
    plt.show()
```

```python
# Function to perform train-test split
def perform_train_test_split(data, test_size=0.2, random_state=42):
    X = data.drop(columns=['ln_denergy','denergy'])
    y = data['ln_denergy']
    return train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python


# Function to standardize data
def standardize_data(X_train, X_test):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    return X_train_scaled, X_test_scaled

# Function to implement regression algorithms and calculate metrics
def implement_regression_algorithms(X_train, y_train, X_test, y_test):
    # Define the parameter grids for hyperparameter tuning


    param_grid_rf = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 15, 20],
    'max_features': ['auto', 'sqrt', 'log2', None],
    'bootstrap': [True, False]
}


    algorithms = {
        'Random Forest': (RandomForestRegressor(), param_grid_rf)
    }

    results = {}

    for name, (model, param_grid) in algorithms.items():
        grid_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, cv=5, n_jobs=-1, scoring='max_error',verbose=3)
        grid_search.fit(X_train, y_train)

        best_model = grid_search.best_estimator_
        y_train_pred = best_model.predict(X_train)
        y_test_pred = best_model.predict(X_test)

        train_metrics = calculate_metrics(y_train, y_train_pred)
        test_metrics = calculate_metrics(y_test, y_test_pred)

        results[name] = {
            'Best Params': grid_search.best_params_,
            'Train Metrics': train_metrics,
            'Test Metrics': test_metrics,
            'Actual Values': y_test,
            'Predicted Values': y_test_pred
        }

    return results

# Function to calculate evaluation metrics
def calculate_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    evs = explained_variance_score(y_true, y_pred)
    me = max_error(y_true, y_pred)
    mae_median = median_absolute_error(y_true, y_pred)

    return {
        'MSE': mse,
        'MAE': mae,
        'R^2': r2,
        'Explained Variance Score': evs,
        'Max Error': me,
        'Median Absolute Error': mae_median
    }
def print_results_in_table(results):
    rows = []
    for name, result in results.items():
        row = {
            'Model': name,
            'Best Params': result['Best Params'],
            'Train Metrics': result['Train Metrics'],
            'Test Metrics': result['Test Metrics'],
        }
        rows.append(row)
    
    # Convert to DataFrame
    df = pd.DataFrame(rows)
    
    # Print DataFrame
    print(df.to_string())
```

### MBTR Descriptor (Many-Body Tensor Representation)

#### What is MBTR Descriptor?
The Many-Body Tensor Representation (MBTR) is a method used in computational chemistry and materials science to describe the geometric and chemical properties of atomic systems. MBTR encodes information about the relative positions and types of atoms in a molecule or material by considering many-body interactions (e.g., pairs, triplets).

#### Importance of MBTR Descriptor
1. **Detailed Geometric Information**: MBTR captures detailed geometric and chemical information about a system, which is crucial for accurately predicting physical and chemical properties.
2. **Scalability**: It can be applied to a variety of atomic systems, from small molecules to complex materials.
3. **Interoperability**: MBTR can be used with various machine learning algorithms, making it a flexible tool for building predictive models in chemistry and materials science.

#### How MBTR Descriptor Works
1. **Atomic Environment Representation**: MBTR considers interactions between atoms, typically up to triplets (three-body interactions).
2. **Distance and Angle Distributions**: The descriptor captures distributions of distances (two-body interactions) and angles (three-body interactions) between atoms.
3. **Weighting Scheme**: Different interactions can be weighted differently based on their chemical significance.
4. **Kernel Density Estimation**: The distributions are smoothed using kernel density estimation to create continuous and differentiable functions.
5. **Tensor Representation**: The smoothed distributions are represented as tensors, forming the final MBTR.

#### Interpretation
MBTR provides a comprehensive and structured representation of an atomic system. Each component of the MBTR tensor corresponds to a specific geometric or chemical interaction (e.g., pairwise distances, angles between triplets). This structured representation can be used as input for machine learning models to predict various properties of the system.

### Example
Consider two molecules: water (H₂O) and hydrogen peroxide (H₂O₂).

1. **Atomic Environment Representation**:
   - For water, consider the interactions between the oxygen atom and the two hydrogen atoms.
   - For hydrogen peroxide, consider interactions between both oxygen atoms and the hydrogen atoms.

2. **Distance and Angle Distributions**:
   - For water, MBTR would capture the O-H distances and the H-O-H angle.
   - For hydrogen peroxide, MBTR would capture O-H distances, O-O distance, and H-O-O-H angles.

3. **Weighting Scheme**:
   - Different interactions (e.g., O-H vs. O-O) can be weighted based on their importance to the molecular properties.

4. **Kernel Density Estimation**:
   - Smooth the distributions of distances and angles to create continuous functions.

5. **Tensor Representation**:
   - Convert the smoothed distributions into tensors, forming the final MBTR for each molecule.

### Advantages of MBTR Descriptor
1. **Comprehensive Representation**: Encodes detailed geometric and chemical information, capturing many-body interactions.
2. **Flexibility**: Applicable to various atomic systems and scalable to complex structures.
3. **Machine Learning Compatibility**: Structured representation suitable for various machine learning algorithms, enhancing prediction accuracy.

### Conclusion
MBTR descriptors are a robust tool for representing atomic systems in computational chemistry and materials science. They provide detailed and structured information about geometric and chemical interactions, making them invaluable for building accurate predictive models. By capturing many-body interactions, MBTR descriptors offer a comprehensive and flexible representation that can be used to predict a wide range of molecular and material properties.




### SOAP Descriptor (Smooth Overlap of Atomic Positions)

#### What is SOAP Descriptor?
The Smooth Overlap of Atomic Positions (SOAP) descriptor is a technique used in computational chemistry and materials science to quantify the similarity between atomic environments. It provides a way to describe local atomic configurations in a smooth, continuous manner, which is highly beneficial for machine learning applications in chemistry.

#### Importance of SOAP Descriptor
1. **Accuracy**: SOAP descriptors offer a high-resolution representation of atomic environments, which can improve the accuracy of predictive models in chemistry and materials science.
2. **Smoothness**: The smooth nature of SOAP descriptors allows for more robust interpolation and extrapolation in machine learning models, leading to better generalization.
3. **Versatility**: SOAP descriptors can be applied to various types of molecular and material data, making them a versatile tool in computational chemistry.

#### How SOAP Descriptor Works
1. **Atomic Environment Representation**: SOAP represents the local environment of an atom by considering the positions of neighboring atoms within a certain cutoff radius.
2. **Gaussian Smeared Density**: The positions of the neighboring atoms are smeared using Gaussian functions to create a smooth atomic density distribution.
3. **Power Spectrum Calculation**: The atomic density is expanded in terms of spherical harmonics and radial basis functions, and the power spectrum of these expansions forms the SOAP descriptor.
4. **Similarity Measure**: The similarity between two atomic environments can be quantified by computing the dot product or kernel between their SOAP descriptors.

#### Interpretation
The SOAP descriptor provides a continuous, high-dimensional representation of the local atomic environment. Each component of the SOAP vector corresponds to a specific feature of the local atomic arrangement, encoded in terms of spherical harmonics and radial basis functions.

### Example
Consider two molecules: methane (CH₄) and ethane (C₂H₆).

1. **Atomic Environment Representation**:
   - For methane, consider the environment around the central carbon atom.
   - For ethane, consider the environment around one of the carbon atoms.

2. **Gaussian Smeared Density**:
   - For methane, Gaussian functions are centered on the four hydrogen atoms around the central carbon.
   - For ethane, Gaussian functions are centered on the hydrogen atoms and the neighboring carbon atom.

3. **Power Spectrum Calculation**:
   - Expand the smeared densities in terms of spherical harmonics and radial basis functions.
   - Calculate the power spectrum for each environment, resulting in the SOAP descriptor.

4. **Similarity Measure**:
   - Compare the SOAP descriptors of methane and ethane to quantify the similarity between their atomic environments.

### Advantages of SOAP Descriptor
1. **High Resolution**: Captures detailed information about the local atomic environment.
2. **Smooth Representation**: Provides a continuous and differentiable representation, beneficial for machine learning.
3. **Versatile**: Can be used for a wide range of chemical and materials science applications.

### Conclusion
SOAP descriptors are a powerful tool in computational chemistry for representing and comparing atomic environments. They provide a high-resolution, smooth, and versatile way to encode local atomic configurations, making them invaluable for machine learning applications in predicting molecular properties and understanding material behavior.


Morgan's fingerprint, in the context of chemistry and cheminformatics, refers to a type of molecular fingerprint used to encode the structure of a molecule for use in various computational analyses, particularly in the context of drug discovery and cheminformatics. It is named after David J. Morgan, who developed the algorithm.

### What is Morgan's Fingerprint?

Morgan's fingerprint is a specific type of circular fingerprint, which encodes the presence of particular substructures (such as atoms and their immediate environment) within a molecule. It is essentially a way to represent the structure of a molecule numerically, making it easier to compare and analyze using computational methods.

### Importance of Morgan's Fingerprint

Morgan's fingerprint is crucial in several areas of computational chemistry, including:

1. **Similarity Searching**: It is used to find molecules that are similar to a given molecule, which is valuable in drug discovery when searching for compounds with similar biological activity.
2. **Quantitative Structure-Activity Relationship (QSAR) Models**: These models predict the activity of compounds based on their chemical structure. Morgan's fingerprints provide a detailed and structured way to represent these molecules.
3. **Clustering and Classification**: In large chemical databases, Morgan's fingerprints help cluster similar molecules together, making it easier to classify and analyze them.

### How It Works

Morgan's algorithm generates fingerprints by considering the environment around each atom in a molecule:

1. **Initialization**: Each atom in the molecule is assigned an initial identifier based on its atomic number and the connectivity (bonds) to neighboring atoms.
2. **Iteration**: In several iterations (often determined by a predefined radius), the identifiers are updated to include information from neighboring atoms. Each iteration expands the radius of consideration, effectively capturing larger substructures around each atom.
3. **Hashing**: The identifiers from all atoms are hashed into a binary or integer array, creating the final fingerprint. Each position in this array represents the presence or absence of specific substructures.

### Advantages of Morgan's Fingerprint

1. **Flexibility**: It captures information about local atomic environments, making it useful for a wide range of molecules.
2. **Efficiency**: The algorithm is computationally efficient, making it suitable for large-scale database searches and analyses.
3. **Resolution**: By adjusting the radius parameter, the level of detail can be controlled, balancing between local and global structural information.

### Interpretation

The binary or integer array produced by Morgan's fingerprint can be interpreted as follows:

- Each bit (in a binary fingerprint) or each integer (in an integer fingerprint) represents the presence of a particular substructure within the molecule.
- Similarity between two molecules can be quantified by comparing their fingerprints, commonly using metrics like Tanimoto similarity.

### Example

Consider two simple molecules: benzene (C₆H₆) and cyclohexane (C₆H₁₂).

1. **Initialization**: Assign each carbon atom an initial identifier.
2. **Iteration**: For radius 1, update each identifier based on its neighbors (for benzene, all are carbon; for cyclohexane, some are hydrogen).
3. **Hashing**: Convert these identifiers into a fingerprint array.

Benzene might have a fingerprint that strongly highlights its aromatic ring structure, while cyclohexane's fingerprint would reflect its saturated ring structure.

### Conclusion

Morgan's fingerprints are a powerful tool in cheminformatics, providing a detailed yet computationally efficient way to represent molecular structures. They are widely used in drug discovery and chemical analysis, enabling the comparison, classification, and prediction of molecular properties.
